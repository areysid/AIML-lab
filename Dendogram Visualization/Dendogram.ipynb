#Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
```

#Loading the Dataset

```python
df = pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")  # Replace with actual file name
```

#Preprocessing Data


```python
df.drop(columns=['CustomerID'], inplace=True)  # Remove ID column
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})  # Encode Gender

# Select relevant features for clustering
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Normalize data for better clustering
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

#Creating Linkage Matrix

```python
linkage_matrix = sch.linkage(features_scaled, method='ward')
```

#Plotting the Dendrogram using Seaborn

```python
plt.figure(figsize=(12, 6))
sns.set(style='whitegrid')
dendrogram = sch.dendrogram(linkage_matrix, no_labels=True)
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.title("Customer Segmentation Dendrogram")
plt.show()
```
